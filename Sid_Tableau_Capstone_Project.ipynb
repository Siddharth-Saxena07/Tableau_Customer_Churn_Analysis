## Task 2 : Identify the Outliers in the dataset.(if there any)

```python
num = data[['CLIENTNUM','Customer_Age','Dependent_count','Months_on_book','Total_Relationship_Count','Months_Inactive_12_mon','Contacts_Count_12_mon','Credit_Limit','Total_Revolving_Bal','Avg_Open_To_Buy','Total_Trans_Amt','Total_Trans_Ct','Avg_Utilization_Ratio']]
num
```

```python
Min = []
Q1 = []
Q2 = []
Q3 = []
Max = []
Range = []
IQR = []
Upper_fence = []
Lower_fence = []
LO = []
UO = []

for col in num:
    
    Min.append(num[col].min())
    q1 = num[col].quantile(0.25)
    Q1.append(q1)
    q2 = num[col].quantile(0.50)
    Q2.append(q2)
    q3 = num[col].quantile(0.75)
    Q3.append(q3)
    Max.append(num[col].max())
    
    Range.append((num[col].max())-(num[col].min()))
    IQR.append(q3-q1)
    Uf = q3+1.5*(q3-q1)
    Upper_fence.append(Uf)
    Lf = q1-1.5*(q3-q1)
    Lower_fence.append(Lf)
    
    for i in num[col]:
        if (i<Lf):
            LO.append(i)
        if (i>Uf):
            UO.append(i)
    
row_head = ['Minimum', '25%', '50%', '75%', 'Maximum', 'Range', 'IQR', 'Upper fence', 'Lower fence', 'Outliers']
col_names = ['CLIENTNUM','Customer_Age','Dependent_count','Months_on_book','Total_Relationship_Count','Months_Inactive_12_mon','Contacts_Count_12_mon','Credit_Limit','Total_Revolving_Bal','Avg_Open_To_Buy','Total_Trans_Amt','Total_Trans_Ct','Avg_Utilization_Ratio']

d_data = pd.DataFrame([Min,Q1,Q2,Q3,Max,Range,IQR,Upper_fence,Lower_fence,[LO,UO]], columns=col_names)
d_data.insert(0,'Measures', row_head)
d_data
```

```python
# separating the categorical and continuos data

def separate_data_types(data):
    categorical = []
    continous = []
    for column in data.columns:
        if data[column].dtypes == object:
            categorical.append(column)
        else:
            continous.append(column)
    return categorical, continous
categorical, continous = separate_data_types(data)

from tabulate import tabulate
table = [categorical, continous]
print(tabulate({'Categorical':categorical, "Continous":continous}, headers = ["categorical" ,"continous"]))
```
